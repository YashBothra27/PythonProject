## IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

DATA MANIPULATING AND CLEANING

DISPLAYING FIRST 10 ROWS FROM THE EXCEL FILE

```python
data = pd.read_csv('vdogame.csv')
print(data.head(10))
```

CREATING A DATAFRAME`

```python
df = pd.DataFrame(data)
print(df)
```

GENERATING INFORMATION OF THE DATAFRAME

```python
print(df.info())
```

FILLING THE NULL VALUES 

```python
df = df.dropna(subset=['Year'])
df['Publisher'] = df['Publisher'].fillna('Unknown')
print(df.info())
```

SORTING COLUMNS

```python
multi_sort = df.sort_values(by=['Year', 'Name'], ascending=[True, False])
print(multi_sort.head(5))
print('______')
print(multi_sort.tail(5))
```

```python
platform_count = df['Platform'].value_counts()
print(f'Total number of platforms: {len(platform_count)}')
print(platform_count.head(5))
print(platform_count.tail(5))               
```

```python
year_count = df['Year'].value_counts()
print(f'Total number of years: {len(year_count)}')
print(year_count.head(5))
print(year_count.tail(5))
```

```python
genre_count = df['Genre'].value_counts()
print(f'Total number of genres: {len(genre_count)}')
print(genre_count.head(5))
print(genre_count.tail(5))
```

```python
publisher_count = df['Publisher'].value_counts()
print(f'Total number of platforms: {len(publisher_count)}')
print(publisher_count.head(5))
print(publisher_count.tail(5))
```

DATA VISUALIZATION USING MATPLOTLIB 

NUMBER OF GAME RELEASED BY GENRE(BAR PLOT)

```python

plt.figure(figsize=(12, 8))
plt.pie(genre_count,labels=genre_count.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Genres',fontsize=16)
plt.legend(genre_count.index, loc='upper right')
plt.show()
```

GAMES RELEASED EVERY YEAR(BAR PLOT)


```python
plt.figure(figsize=(10, 6))
plt.bar(year_count.index, year_count.values, color='skyblue')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Number of Games',fontsize=12)
plt.title('Number of Games Released by Year',fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

TOTAL GLOBAL SALES(LINE PLOT)

```python
sales_yearly = data.groupby('Year')['Global_Sales(mil)'].sum()

plt.figure(figsize=(10, 6))
plt.plot(sales_yearly.index, sales_yearly.values, marker='o', linestyle='-', color='green')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Global Sales (Million)', fontsize=12)
plt.title('Total Global Sales of Video Games every Year', fontsize=16)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

GLOBAL SALES BY GENRE(HORIZONTAL BAR PLOT)

```python
sales_by_genre = df.groupby('Genre')['Global_Sales(mil)'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 7))
plt.barh(sales_by_genre.index, sales_by_genre.values, color='red')
plt.xlabel('Global sales(millions)',fontsize=12)
plt.ylabel('Genre',fontsize=12)
plt.title('Total Global Sales by Genre',fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
platform_games = data['Platform'].value_counts()

plt.figure(figsize=(12, 7))
plt.bar(platform_games.index, platform_games.values, color='purple')
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Number of Games', fontsize=12)
plt.title('Number of Games Released in each platform', fontsize=16)
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```
